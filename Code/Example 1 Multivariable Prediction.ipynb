```python
from keras import layers
```

```python
from keras import models
```

```python
import numpy as np
```

```python
import math
```

```python
training_data = np.array([[-3,-6,-2],[-5,-8,-8],[-7,-2,-7],[-1,-2,-9],[-3,-4,-6],[-5,-7,-1],[3,6,2],[5,8,8],[7,2,7],[1,2,9],[3,4,6],[5,7,1]],"float32")
```

```python
validation_data = np.array([[-52],[-118],[-104],[-84],[-78],[-59],[52],[118],[104],[84],[78],[59]],"float32")
```

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(1, input_dim=3))
```

```python
Image(url="https://live.staticflickr.com/65535/48237211596_f35775f553_b.jpg", width = 500 , height = 500)
```

```python
model.compile(loss='mse', optimizer='sgd',metrics=['mse'])
```

```python
history = model.fit(training_data, validation_data , epochs=40 , verbose=2)
```

```python
print(model.layers[0].get_weights()[0]) # weights 
print(model.layers[0].get_weights()[1]) # bias 
```

```python
testing_data = np.array([[1,1,1]])
```

```python
print(model.predict(testing_data))
```

```python
from matplotlib import pyplot
```

```python
history.history.keys()
```

```python
pyplot.plot(history.history['mse'])
```

Try to compile it using other optimizer like Adam Optimizer , you will figure out something

```python
model2 = models.Sequential()
```

```python
model2.add(layers.Dense(1, input_dim=3))
```

```python
model2.compile(loss='mse', optimizer='adam',metrics=['mse'])
```

```python
history2 = model2.fit(training_data, validation_data , epochs=8000 , verbose=2)
```

```python
pyplot.plot(history2.history['mse'])
```

```python
print(model2.layers[0].get_weights()[0]) # weights 
print(model2.layers[0].get_weights()[1]) # bias 
```
