```python
from keras import layers
```

```python
from keras import models
```

```python
import numpy as np
```

```python
import math
```

```python
# training_data = np.array([[3,6,2],[5,8,8],[7,2,7],[1,2,9],[3,4,6],[5,7,1]],"float32")
training_data = np.array([[1,1,6],[2,2,5],[3,3,4],[4,4,3],[5,5,2],[6,6,1]],"float32")
```

```python
# validation_data = np.array([[52],[118],[104],[84],[78],[59]],"float32")
validation_data = np.array([[27],[26],[25],[24],[23],[22]],"float32")
```

```python
#For The model with bias
model = models.Sequential()
model.add(layers.Dense(1, input_dim=3,  use_bias=True))
```

```python
Image(url="https://live.staticflickr.com/65535/51308762083_d2d0fc35d2_o.png", width = 300 , height = 300)
```

```python
# For The model without bias
# model.add(layers.Dense(1, input_dim=3,  use_bias=None)) 
```

```python
Image(url="https://live.staticflickr.com/65535/48237211596_f35775f553_b.jpg", width = 300 , height = 300)
```

```python
model.compile(loss='mse', optimizer='sgd',metrics=['mse'])
```

```python
history = model.fit(training_data, validation_data , epochs=40 , verbose=2)
```

```python
print(model.layers[0].get_weights()[0]) # weights 
if len(model.layers[0].get_weights()) > 1:
    print(model.layers[0].get_weights()[1]) # bias 
```

```python
testing_data = np.array([[1,1,1]])
```

```python
print(model.predict(testing_data))
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(history.history['mse'])
```
