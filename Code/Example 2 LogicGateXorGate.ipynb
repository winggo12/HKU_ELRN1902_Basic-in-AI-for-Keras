Example 1 : Logic Gate - Xor Gate

```python
from IPython.display import Image
from IPython.core.display import HTML
Image(url= "http://www.vlsiinterviewquestions.org/wp-content/uploads/2012/04/xor.jpg")
```

```python
from keras import layers
```

```python
from keras import models
```

```python
import numpy as np
```

```python
import math
```

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]],"float32")
```

```python
validation_data = np.array([[0],[1],[1],[0]],"float32")
```

```python
model = models.Sequential()
```

```python
layer_input2hidden = layers.Dense(4, input_dim=2, activation='relu')
#layer_input2hidden = layers.Dense(4, input_dim=2,activation='linear')
```

```python
model.add(layer_input2hidden) #input_shape=(2,)
```

```python
layer_hidden2output = layers.Dense(1 , activation='sigmoid')
```

```python
model.add(layer_hidden2output)
```

```python
Image(url='https://blog.thoughtram.io/images/xor_model.png')
```

```python
model.compile(loss='mse', optimizer='adam',metrics=['binary_accuracy','mse','acc'])
```

```python
history = model.fit(training_data, validation_data , epochs=5000 , verbose=2) #verbose=0 --> No log 
```

Finding the weights and bias in the input-to-hidden layer:

```python
layer_input2hidden.get_weights()
```

The lines of code above and below have the same meaning , which output the weights and bias in the input-to-hidden layer

```python
print(model.layers[0].get_weights()[0]) # weights in the input-to-hidden layer
print(model.layers[0].get_weights()[1]) # bias in the input-to-hidden layer
```

Finding the weights and bias in the hidden_to_output layer:

```python
layer_hidden2output.get_weights()
```

```python
print(model.layers[1].get_weights()[0]) # weights in the hidden-to-output layer
print(model.layers[1].get_weights()[1]) # bias in the hidden-to-output layer
```

How to get the result on your own ?

```python
prediction = model.predict(testing_data)
print(prediction)
```

```python
rounded_prediction = prediction.round()
print(rounded_prediction)
```

Defining a Relu and Sigmoid Function

```python
def Relu(x):

    final_x = np.maximum(0,x)
        
    return final_x
```

```python
def Sig(x):
    
    final_x = 1 / (1 + math.e**-x)
    
    return final_x
```

How about calculating the result using mathematical equation?

```python
output_of_hidden_layer = Relu(np.dot(testing_data , model.layers[0].get_weights()[0]) + model.layers[0].get_weights()[1])
print(output_of_hidden_layer)
```

```python
final_output = Sig(np.dot(output_of_hidden_layer ,  model.layers[1].get_weights()[0]) + model.layers[1].get_weights()[1])
print(final_output)
```
